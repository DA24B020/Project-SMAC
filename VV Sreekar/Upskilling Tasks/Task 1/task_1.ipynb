### Question-2

```python
x = np.random.randint(1, 101, size=20)
y = np.random.randint(1, 101, size=20)

dist_from_origin = {}
for i in range(20):
    dist_from_origin[i] = x[i]**2 + y[i]**2

sorted_dist_from_origin = dict(sorted(dist_from_origin.items(), key=lambda item: item[1]))

for j, (key, val) in enumerate(sorted_dist_from_origin.items()):
    for k in range(20):
        dist_from_origin[k] = (x[k] - x[j])**2 + (y[k] - y[j])**2
    sorted_dist_from_point = dict(sorted(dist_from_origin.items(), key=lambda item: item[1]))
    print(f'point: {key + 1}, co-ordinates: {(x[key], y[key])}')
    print('nearest 3 points - ', end = ' ')
    for t in range(3):
        if (t != 2):
            print(list(sorted_dist_from_point.items())[t + 1][0] + 1, end = ', ')
        else:
            print(list(sorted_dist_from_point.items())[t + 1][0] + 1)
        x1, x2 = x[j], x[list(sorted_dist_from_point.items())[t + 1][0]]
        y1, y2 = y[j], y[list(sorted_dist_from_point.items())[t + 1][0]]
        plt.plot((x1, x2), (y1, y2))
    
```

### Question-3

```python
A = np.array([[0, 0, 0, 0, -170], [1, 0, 0, 0, -10], [0, 1, 0, 0, 1], [0, 0, 1, 0, 13], [0, 0, 0, 1, 0]])

theta = np.linspace(0, 6.28)

radii = [11, 2, 14, 1]
for i in range(len(radii)):
    x = radii[i] * np.sin(theta)
    y = radii[i] * np.cos(theta)
    plt.plot(x, y)

eig_vals, eig_vects = np.linalg.eig(A)

for j in range(5):
    plt.plot(eig_vals[j].real, eig_vals[j].imag, 'o')
print(eig_vals)
```

### Question-4

```python
from scipy.interpolate import CubicSpline

x_data = np.array([0, 1, 2, 3, 4, 5, 9])
y_data = np.array([2.5, 0.5, 0.5, 2, 2, 1.5, 0])

# natural cubic spline interploation
cs = CubicSpline(x_data, y_data, bc_type='natural')

x= np.linspace(x_data.min(), x_data.max(), 500)
y = cs(x)

plt.figure(figsize=(8, 5))
plt.plot(x_data, y_data, 'o', label='Data Points')
plt.plot(x, y, label='Natural Cubic Spline')
plt.grid(True)

# polynomial interpolation
degree = len(x_data) - 1
coeffs = np.polyfit(x_data, y_data, degree)
poly = np.poly1d(coeffs)

x = np.linspace(x_data.min(), x_data.max(), 500)
y = poly(x)

plt.plot(x, y, label=f'Interpolating Polynomial (deg={degree})')
plt.legend()
plt.title("Natural Cubic Interploation Vs Polynomial Interpolation")
plt.grid(True)
plt.show()

```
